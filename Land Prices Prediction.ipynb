```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("aqardata_2.csv")
df.head(1)
```

```python
df.info()
```

```python
df_clean = df.dropna(how='any',axis=0)
```

```python
df_reindexed = df_clean.reset_index(drop=True)
```

```python
df_reindexed.info()
```

```python
df_reindexed.head(20)
```

```python
numeric_columns = ["streetwidth" , "size"]
nominal_columns = ["mainlocation" , "sublocation" , "neighborhood","frontage","purpose"]
```

```python
df_dummy = pd.get_dummies(df_reindexed)
df_dummy
```

```python
# from sklearn.preprocessing import MinMaxScaler

# scaler = MinMaxScaler()

# df_dummy[numeric_columns] = scaler.fit_transform(df_dummy[numeric_columns])
# df_dummy.head()
df_dummy.columns
```

```python
X = df_dummy.drop(["Pricepm"], axis=1)
y = df_dummy["Pricepm"]

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
models = {
    "LinearRegression":LinearRegression(),
    "Ridge":Ridge()
}
```

```python
for model_name in models:
    model = models[model_name]
    results = cross_validate(model, X, y, cv=10, scoring=['neg_mean_absolute_error', 'neg_root_mean_squared_error'
                                                        ,'r2'], return_train_score=True)

    print(model_name + ": ")
    print("Mean Absolute Error:" , 'train: ', abs(results['train_neg_mean_absolute_error'].mean()),
          '| test: ', abs(results['test_neg_mean_absolute_error'].mean()))
    print("Root Mean Square Error:" , 'train: ', abs(results['train_neg_root_mean_squared_error'].mean()),
          '| test: ', abs(results['test_neg_root_mean_squared_error'].mean()))
    print("R2:" , 'train: ', results['train_r2'].mean(),
          '| test: ', results['test_r2'].mean())
    print("---------------------------------------------------------")
```

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, 
                                                    random_state = 5)
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
```

Ridge model is producing better results than the LinearRegression

```python
mod = Ridge()
mod.fit(X_train,y_train)
```
