# 1- importing the necessary libraries

```python
import pandas as pd
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

# 2- Gathering, cleaning, assessing data 

```python
df = pd.read_csv("aqardata_2.csv")
df.head(1)
```

```python
df.info()
```

```python
df_clean = df.dropna(how='any',axis=0)
```

```python
df_reindexed = df_clean.reset_index(drop=True)
```

```python
df_reindexed.info()
```

```python
df_reindexed.head()
```

```python
numeric_columns = ["streetwidth" , "size"]
nominal_columns = ["mainlocation" , "sublocation" , "neighborhood","frontage","purpose"]
```

#3- Preprocessing

```python
df_dummy = pd.get_dummies(df_reindexed)
df_dummy
```

```python
X = df_dummy.drop(["Pricepm"], axis=1)
y = df_dummy["Pricepm"]

print(X.shape)
print(y.shape)
```

# 4- Training and Evaluation Result

```python
models = {
    "LinearRegression":LinearRegression(),
    "Ridge":Ridge()
}
```

```python
for model_name in models:
    model = models[model_name]
    results = cross_validate(model, X, y, cv=10, scoring=['neg_mean_absolute_error', 'neg_root_mean_squared_error'
                                                        ,'r2'], return_train_score=True)

    print(model_name + ": ")
    print("Mean Absolute Error:" , 'train: ', abs(results['train_neg_mean_absolute_error'].mean()),
          '| test: ', abs(results['test_neg_mean_absolute_error'].mean()))
    print("Root Mean Square Error:" , 'train: ', abs(results['train_neg_root_mean_squared_error'].mean()),
          '| test: ', abs(results['test_neg_root_mean_squared_error'].mean()))
    print("R2:" , 'train: ', results['train_r2'].mean(),
          '| test: ', results['test_r2'].mean())
    print("---------------------------------------------------------")
```
